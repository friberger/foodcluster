```python
import numpy as np
import csv
```

```python
def unicode_csv_reader(utf8_data, dialect=csv.excel, **kwargs):
        csv_reader = csv.reader(utf8_data, dialect=dialect, **kwargs, delimiter = ';')
        for row in csv_reader:
            yield row
```

```python
data = []
filename = 'Huvudgrupp171130_LDB_ppen_kod5.csv'
reader = unicode_csv_reader(open(filename))
for field1 in reader:
    data.append(field1)

reader.close()

data
```
